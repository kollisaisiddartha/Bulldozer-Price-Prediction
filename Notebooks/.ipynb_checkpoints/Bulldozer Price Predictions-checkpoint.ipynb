```python
#filling numerical values

#checking which numeric columns has null values
for label,content in df_tmp.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            df_tmp[label+'_is_missin'] = pd.isnull(content) #adding binary column to check data missin or not
            df_tmp[label] = content.fillna(content.median())
```

```python
#filling categorical values and changing all categories to nums at same time
for label, content in df_tmp.items():
    if not pd.api.types.is_numeric_dtype(content):
        df_tmp[label+'_is_missing'] = pd.isnull(content)
        df_tmp[label] = pd.Categorical(content).codes+1 # 1 is added because in codes if there are null values pandas assign -1
```

```python
#splitting our data into train and validation sets
df_val = df_tmp[df_tmp.saleYear==2012]
df_train = df_tmp[df_tmp.saleYear != 2012]

len(df_val) , len(df_train)
```

```python
#split data into x and y
X_train, y_train = df_train.drop('SalePrice', axis=1), df_train.SalePrice
X_valid, y_valid = df_val.drop('SalePrice', axis=1), df_val.SalePrice

X_train.shape , y_train.shape , X_valid.shape , y_valid.shape
```

```python
#building an evaluation fucntion
from sklearn.metrics import mean_squared_log_error, mean_absolute_error, r2_score

def rmsle(y_test, y_preds):
    return np.sqrt(mean_squared_log_error(y_test, y_preds))

def show_scores(model):
    train_preds = model.predict(X_train)
    valid_preds = model.predict(X_valid)
    scores = {'Training MAE':mean_absolute_error(y_train, train_preds),
              'Valid MAE':mean_absolute_error(y_valid, valid_preds),
              'Training RMSLE':rmsle(y_train, train_preds),
              'Valid RMSLE':rmsle(y_valid, valid_preds),
              'Training R^2':r2_score(y_train, train_preds),
              'Valid R^2':r2_score(y_valid, valid_preds)}
    return scores
```

```python
#reducing data using MAX SAMPLES and testing for tuning
model = RandomForestRegressor(n_jobs=-1,
                              random_state=69,
                              max_samples=10000)
```

```python
%%time
model.fit(X_train, y_train)
```

```python
show_scores(model)
```

```python
#hyperparamter tuning with randamizedsearchcv
from sklearn.model_selection import RandomizedSearchCV

rf_grid = {'n_estimators':np.arange(10,100,10),
           'max_depth':[None,3,5,10],
           'min_samples_split':np.arange(2,20,2),
           'min_samples_leaf':np.arange(1,20,2),
           'max_features':[0.5, 1, 'sqrt', 'auto'],
           'max_samples':[10000]}

rs_model = RandomizedSearchCV(RandomForestRegressor(n_jobs=-1,
                                                    random_state=69),
                              param_distributions=rf_grid,
                              n_iter=5,
                              cv=5,
                              verbose=True)

rs_model.fit(X_train, y_train)
```

```python
rs_model.best_params_
```

```python
show_scores(rs_model)
```

```python
%%time
#BEST HYPERPARAMETERS

ideal_model = RandomForestRegressor(n_estimators=40,
                                    min_samples_leaf=1,
                                    min_samples_split=14,
                                    max_features=0.5,
                                    n_jobs=-1,
                                    max_samples=None,
                                    random_state=69)

ideal_model.fit(X_train, y_train)
```

```python
show_scores(ideal_model) #ideal model trained on all data
```

```python
show_scores(rs_model) #model trained on 10000 rows data
```

```python
#preprocess the test df

def preprocess_data(df):
    
    df['saleYear'] = df.saledate.dt.year
    df['saleMonth'] = df.saledate.dt.month
    df['saleDay'] = df.saledate.dt.day
    df['saleDayofWeek'] = df.saledate.dt.dayofweek
    df['saleDayofYear'] = df.saledate.dt.dayofyear
    
    df.drop('saledate', axis=1, inplace=True)
    
    #fill numeric rows with median
    for label,content in df.items():
        if pd.api.types.is_numeric_dtype(content):
            if pd.isnull(content).sum():
                df[label+'_is_missin'] = pd.isnull(content) #adding binary column to check data missin or not
                df[label] = content.fillna(content.median())
         
    #fill categorical missinng values and turn into nums
        if not pd.api.types.is_numeric_dtype(content):
            df[label+'_is_missing'] = pd.isnull(content)
            df[label] = pd.Categorical(content).codes+1
                
    return df
```

```python
def plot_features(columns, importances, n=20):
    df = (pd.DataFrame({'features':columns,
                       'feature_importances':importances})
         .sort_values('feature_importances', ascending=False)
         .reset_index(drop=True))
    
    #plot
    fig, ax = plt.subplots()
    ax.barh(df['features'][:n],df['feature_importances'][:n])
    ax.set_ylabel('Features')
    ax.set_xlabel('Feature Importance')
    ax.invert_yaxis()
```

```python
plot_features(X_train.columns, ideal_model.feature_importances_)
```
